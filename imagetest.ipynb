```python
import pdfkit
import pandas as pd

df = pd.read_csv('./pull_requests_summary.csv')
html_table = df.to_html()

options = {    'page-size': 'Letter',
   'margin-top': '0mm',
   'margin-right': '0mm',
   'margin-bottom': '0mm',
   'margin-left': '0mm'
}

#pdfkit.configuration(wkhtmltopdf='D:/htmltopdf/wkhtmltopdf/bin')
pdfkit.from_string(html_table, 'outputs.pdf', options=options)
```
